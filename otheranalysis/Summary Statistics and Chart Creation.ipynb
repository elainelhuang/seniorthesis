# Factor Plots

```python
# Extract the monthly GMB factors
copy = factordata.copy()
GMBs = (copy.drop_duplicates(subset='GMB_U', keep='first')).copy()
GMBs['date'] = pd.to_datetime(GMBs[['year', 'Month']].assign(day=1))
GMBs = GMBs.sort_values(['date'], ascending=[True])
```

```python
# Converting column data into numpy arrays for easier use in plotting
dates = GMBs['date'].to_numpy()
GMB_Us = GMBs['GMB_U'].to_numpy()
GMB_Ss = GMBs['GMB_S'].to_numpy()
MktRFs = GMBs['Mkt-RF'].to_numpy()
SMBs = GMBs['SMB'].to_numpy()
HMLs = GMBs['HML'].to_numpy()
RMWs = GMBs['RMW'].to_numpy()
CMAs = GMBs['CMA'].to_numpy()
```

## Trailing Twelve Month (TTM) Rolling Average Plot

```python
# Calculate the TTM rolling averages of numpy arrays
def rolling_avg_last_twelve(arr):
    if len(arr) < 12:
        raise ValueError("Array length must be at least 12 for a 12-period rolling average.")
    
    ret = np.convolve(arr, np.ones(12), 'valid') / 12
    return ret
```

```python
rolling_GMB_Us = rolling_avg_last_twelve(GMB_Us)
rolling_GMB_Ss = rolling_avg_last_twelve(GMB_Ss)
rolling_MktRFs = rolling_avg_last_twelve(MktRFs)
rolling_SMBs = rolling_avg_last_twelve(SMBs)
rolling_HMLs = rolling_avg_last_twelve(HMLs)
rolling_RMWs = rolling_avg_last_twelve(RMWs)
rolling_CMAs = rolling_avg_last_twelve(CMAs)
```

```python
# Create subplots
fig, axes = plt.subplots(2, 1, figsize=(12, 9), sharey=True)
TTM_dates = dates[11:]

ax1 = axes[0]
ax1.plot(TTM_dates, rolling_MktRFs, label='TTM Mkt-RF', color='cornflowerblue', alpha=0.45)
ax1.plot(TTM_dates, rolling_SMBs, label='TTM SMB', color='darkorange', alpha=0.45)
ax1.plot(TTM_dates, rolling_HMLs, label='TTM HML', color='firebrick', alpha=0.45)
ax1.plot(TTM_dates, rolling_RMWs, label='TTM RMW', color='rebeccapurple', alpha=0.45)
ax1.plot(TTM_dates, rolling_CMAs, label='TTM CMA', color='goldenrod', alpha=0.45)
ax1.plot(TTM_dates, rolling_GMB_Us, label='TTM GMB$_U$', color='mediumseagreen', linewidth=2.5)
ax1.set_title('TTM Rolling Averages of GMB$_U$ and Fama-French Factors', size=16)
ax1.grid(True)
ax1.legend(ncol=6, loc='lower center')

ax2 = axes[1]
ax2.plot(TTM_dates, rolling_MktRFs, label='TTM Mkt-RF', color='cornflowerblue', alpha=0.45)
ax2.plot(TTM_dates, rolling_SMBs, label='TTM SMB', color='darkorange', alpha=0.45)
ax2.plot(TTM_dates, rolling_HMLs, label='TTM HML', color='firebrick', alpha=0.45)
ax2.plot(TTM_dates, rolling_RMWs, label='TTM RMW', color='rebeccapurple', alpha=0.45)
ax2.plot(TTM_dates, rolling_CMAs, label='TTM CMA', color='goldenrod', alpha=0.45)
ax2.plot(TTM_dates, rolling_GMB_Ss, label='TTM GMB$_S$', color='darkgreen', linewidth=2.5)
ax2.set_title('TTM Rolling Averages of GMB$_S$ and Fama-French Factors', size=16)
ax2.grid(True)
ax2.legend(ncol=6, loc='lower center')

plt.tight_layout()

# Save the plot as PDF
plt.savefig('GMB_FF_Factors_TTM.pdf', dpi=300, bbox_inches='tight')

# Display the plot
plt.show()
```
